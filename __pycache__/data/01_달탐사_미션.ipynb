```python
import pandas as pd

# 데이터프레임 생성을 위한 데이터를 딕셔너리 형태로 만듭니다.
data = {
    'Mission': ['artemis1', 'artemis1b', 'artemis2'],
    'Total weight (kg)': [26520, 26520, 26520],
    'Payload (kg)': [26988, 37965, 42955],
    'Sample weight from total (kg)': [75.549232, 75.549232, 75.549232],
    'Sample weight from payload (kg)': [38.779584, 54.552649, 61.722877]
}

# 딕셔너리를 사용하여 데이터프레임을 생성합니다.
artemis_mission = pd.DataFrame(data)

# 생성된 데이터프레임을 출력합니다.
artemis_mission
```
